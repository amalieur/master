```python
# Cell to create a testplot over the testset in Porto
DATA_FOLDER = "../data\chosen_data\porto\\"
PORTO_META_TEST = "../data\chosen_data\porto\META-1000.txt"

files = read_meta_file(PORTO_META_TEST)

trajectories = load_trajectory_files(files, DATA_FOLDER)

N = len(trajectories)
cmap = plt.get_cmap('gist_ncar')
fig = plt.figure()
fig.set_size_inches(10,8)
for i, key in enumerate(trajectories.keys()):
    #if key not in ["P_ASV", "P_ADH" ]:#"P_AEQ", "P_ABZ"]: #, "P_ADH", "P_ABK""P_ARM"]:
    #    continue
    values = trajectories[key]
    lats, lons = list(zip(*values))

    color = cmap(float(i)/N)

    plt.plot(lons, lats, c=color, zorder=-1)

plt.show()
    
    
```

```python
# Testcell to create a plot over the testset in Rome
DATA_FOLDER = "../data\chosen_data\\rome\\"
PORTO_META_TEST = "../data\chosen_data\\rome\META-1000.txt"

files = read_meta_file(PORTO_META_TEST)

trajectories = load_trajectory_files(files, DATA_FOLDER)


N = len(trajectories)
cmap = plt.get_cmap('gist_rainbow')
fig = plt.figure()
fig.set_size_inches(10,8)
for i, key in enumerate(trajectories.keys()):
    values = trajectories[key]
    lats, lons = list(zip(*values))

    color = cmap(float(i)/N)

    plt.plot(lons, lats, c=color, zorder=-i)

plt.show()
```
