```python
# Cell to create a testplot over the testset in Porto
DATA_FOLDER = "../data/chosen_data/porto/"
PORTO_META_TEST = "../data/chosen_data/porto/META-1000.txt"

files = read_meta_file(PORTO_META_TEST)

trajectories = load_trajectory_files(files, DATA_FOLDER)

N = len(trajectories)

plt.rcParams["font.size"] = 16
cmap = plt.get_cmap('Wistia')
fig = plt.figure(dpi=300)
fig.set_size_inches(10,8)
for i, key in enumerate(trajectories.keys()):
    #if key not in ["P_AAA", "P_AAC","P_AAL", "P_AAP", "P_AAW", "P_ABA", "P_ABG", "P_ABK","P_ABL","P_ABN", "P_ABT","P_ABY","P_ACA", "P_ACB"]:#"P_AEQ", "P_ABZ"]: #, "P_ADH", "P_ABK""P_ARM"]:
    #    continue
    values = trajectories[key]
    lats, lons = list(zip(*values))

    color = cmap(0.65)

    plt.plot(lons, lats, c=color, zorder=i)

plt.show()
    
    
```
